```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from bs4 import BeautifulSoup
```

```python
# Open the file and load the data
with open('queries.json', 'r') as f:
    data = json.load(f) 
def fun(n):
    try:
        query = data[n]
        # Extract the relevant fields from the JSON object
        patient_context = query["profile_context"]
        latest_query = "\n".join([f"{q['role']}: {q['content']}" for q in query['latest_query']])
        ideal_response = query["ideal_response"]
        diet_chart = patient_context["diet_chart"]
        try:
            notes_html = diet_chart.get("notes", "")
            soup = BeautifulSoup(notes_html, "lxml")
            formatted_notes = soup.get_text(separator="\n").strip()
            diet_chart["notes"] = formatted_notes
        except:
            pass

        # Example of accessing meals by day
        meals_by_days = diet_chart.get("meals_by_days", [])
        for day in meals_by_days:
            print(f"Day order: {day['order']}")
            meals = day.get("meals", [])
            for meal in meals:
                print(f"  Meal ID: {meal['id']}")
                print(f"  Meal Name: {meal['name']}")
                print(f"  Meal Timings: {meal['timings']}")
                meal_options = meal.get("meal_options", [])
                for option in meal_options:
                    print(f"    Meal Option ID: {option['id']}")
                    print(f"    Meal Option Notes: {option['notes']}")
                    food_items = option.get("meal_option_food_items", [])
                    for food_item in food_items:
                        print(f"      Food ID: {food_item['food_id']}")
                        print(f"      Food Measure ID: {food_item['food_measure_id']}")
    except Exception as e:
        raise e
```
